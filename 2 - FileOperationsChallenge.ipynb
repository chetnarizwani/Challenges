```python
import os

os.chdir('C:\\Users\\YOGITA\\Desktop\\Python\\iNeuron\\Python\\Assignments\\Questions')

path = os.getcwd()

#Renaming of files
for file in os.listdir():
    if file.endswith("(1).docx") :
        fileno = []
        fileno_str = ""
        for i in file:
            if i.isdigit():
                fileno.append(i)
            elif i == "(":
                break
            fileno_str = "".join(fileno)
        os.rename(path+"\\"+file,path+"\\"+file.replace("(1).docx","")+"Adv.docx")

#Find common filenames
import os
remove_list = []
remove_text = []
dir1 = os.listdir()
i = 0
j = 0
for file in dir1:
    remove_list.append(file.replace(".docx",""))

while i < len(remove_list) - 1:
    small = remove_list[i] if len(remove_list[i]) < len(remove_list[i+1]) else remove_list[i+1]
    for j in range(len(small)):
        if remove_list[i][j] != remove_list[i+1][j]:
            if remove_list[i][0:j]!="":
                remove_text.append(remove_list[i][0:j])
            break
    i += 1
print(remove_text)

#Create directories
for i in remove_text:
    os.mkdir(i)

#Move files to respective directories    
import shutil
for file in os.listdir():
    if file.endswith(".docx"):
        for common in remove_text:
            if file.startswith(common):
                shutil.move(file,path+"\\"+common)

import docx
os.chdir('C:\\Users\\YOGITA\\Desktop\\Python\\iNeuron\\Python\\Assignments\\Questions')

#Create compiled document
for i in remove_text:
    os.chdir(os.getcwd() + "\\"+i)
    fwrite = docx.Document()
    for j in range(len(os.listdir())):
        if os.listdir()[j].startswith(i):
            file = os.listdir()[j]
            doc = docx.Document(file)
            fulltext = []
            for para in doc.paragraphs:
                fulltext.append(para.text)
            fwrite.add_heading(file,0)
            fwrite.add_paragraph('\n'.join(fulltext))
            fwrite.save(os.getcwd()+"\\"+i+"_Compiled.docx")
    os.chdir('C:\\Users\\YOGITA\\Desktop\\Python\\iNeuron\\Python\\Assignments\\Questions') 

print("Done")
```

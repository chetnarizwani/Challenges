```python
import tkinter as tk
from tkinter.filedialog import askopenfilename
from PyPDF2 import PdfFileMerger, PdfFileReader
from pathlib import Path
import docx
from tkinter import messagebox

filelist = []

# initiate PDF merger Object
merger = PdfFileMerger()

#inititate doc merger object
docmerger = docx.Document()

def open_file(files):
    if variable.get() == ".pdf":
        filepath = askopenfilename(
            filetypes=[("PDF Files","*.pdf"), ("All Files", "*.*")]
        )
        
    elif variable.get() == ".txt":
        filepath = askopenfilename(
        filetypes=[("Text Files","*.txt"), ("All Files", "*.*")]
        )
       
    elif variable.get() == ".docx":
        filepath = askopenfilename(
        filetypes=[("All Word Documents","*.docx"), ("All Files", "*.*")]
        )
        
    if not(filepath and Path(filepath).exists()):
            return
    files.append(filepath)
    lbl_items["text"] = '\n'.join(str(f) for f in files)
    if len(files) >= 2 and btn_merge['state'] == "disabled":
        btn_merge["state"] = "normal"
        

def merge_files(files):
    if variable.get() == ".pdf":
        for f in files:
            merger.append(PdfFileReader(open(f, "rb")))
        output_filename = ent_output_name.get()
        if not output_filename:
            output_filename = "Untitled.pdf"
        elif ".pdf" not in output_filename:
            output_filename += ".pdf"
        merger.write(output_filename)
        
    elif variable.get() == ".txt":
        filenames = files
        output_filename = ent_output_name.get()
        if not output_filename:
            output_filename = "Untitled.txt"
        elif ".txt" not in output_filename:
            output_filename += ".txt"
        with open(output_filename, 'w') as outfile:
            for names in filenames:
                with open(names) as infile:
                    outfile.write(infile.read())
                outfile.write("\n")
                
    elif variable.get() == ".docx":
        for file in files:
            doc = docx.Document(file)
            fulltext = []
            for para in doc.paragraphs:
                fulltext.append(para.text)
            docmerger.add_heading(file,0)
            docmerger.add_paragraph('\n'.join(fulltext))
            output_filename = ent_output_name.get()
            if not output_filename:
                output_filename = "Untitled.docx"
            elif ".docx" not in output_filename:
                output_filename += ".docx"
            docmerger.save(output_filename)   
    messagebox.showinfo("Success", "Files merged! Please find the compiled file in your current working directory. Thank You!")

# create desktop GUI
window = tk.Tk()
window.title("File Merger")
# window.geometry("500x300")
ws = window.winfo_screenwidth()
hs = windo.winfo_screenheight()
# calculate position x, y
x = (ws/2) - (500/2)    
y = (hs/2) - (300/2)
window.geometry('%dx%d+%d+%d' % (500, 300, x, y))
# not allowed resizing x y direction
window.resizable(0,0)

#create drop down
OptionList = [
".txt",
".pdf",
".docx",
] 
variable = tk.StringVar(window)
variable.set(OptionList[0])

opt = tk.OptionMenu(window, variable, *OptionList)
opt.config(width=90, font=('Helvetica', 12))
opt.pack(side="top")
labelTest = tk.Label(text="", font=('Helvetica', 10))

labelTest.pack(side="top")

def callback(*args):
    labelTest.configure(text="The selected file type to merge is {}".format(variable.get()))
    lbl_items["text"] = ""
    filelist.clear()
variable.trace("w", callback)


# --- Ask open files ---
fr_bg1 = tk.Frame(window, bd=3)
lbl_open = tk.Label(fr_bg1, text="Please choose files to join: (2 and above)")
lbl_open.grid(row=0, column=0, sticky="ew", padx=5, pady=5)

btn_open = tk.Button(fr_bg1, text="Open file(s)",
                command=lambda: open_file(filelist))
btn_open.grid(row=1, column=0, sticky="ew", padx=5)
lbl_items = tk.Label(fr_bg1, text="")
lbl_items.grid(row=2, column=0, pady=5)
fr_bg1.pack()

# --- Button to merge files ---
fr_bg2 = tk.Frame(window, bd=3)
lbl_to_merge = tk.Label(fr_bg2, text="Merge selected files)")
lbl_to_merge.grid(row=0, column=0, sticky="ew", padx="5", pady="5")

ent_output_name = tk.Entry(master=fr_bg2, width=7)
ent_output_name.grid(row=1, column=0, sticky="ew")

btn_merge = tk.Button(fr_bg2, 
                text="Merge files",
                state="disabled",
                command=lambda: merge_files(filelist))
btn_merge.grid(row=2, column=0, sticky="ew", padx=5, pady=5)
fr_bg2.pack()

# --- Button to exit ---
btn_exit = tk.Button(window, text="Exit", command=window.destroy, bd=2)
btn_exit.pack(side=tk.BOTTOM, fill=tk.BOTH, expand=tk.FALSE)

if __name__ == "__main__":
    window.mainloop()
```
